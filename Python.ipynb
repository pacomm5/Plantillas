##68. Flatten a list using Numpy and itertools. esto quiere decir convertir una lista de varias. 

```python
import itertools
L1 = [[1, 2, 3, 4,20,30,40], [5, 6, 7,99,97,98], [81, 9, 100]]
flat_list = list(itertools.chain(*L1))
print(sorted(flat_list))
[1, 2, 3, 4, 5, 6, 7, 9, 20, 30, 40, 81, 97, 98, 99, 100]

import numpy
L1 = [[1, 2, 3, 4,20,30,40], [5, 6, 7,99,97,98], [81, 9, 100]]
flat_list = list(numpy.concatenate(L1).flat)
print(sorted(flat_list))
[1, 2, 3, 4, 5, 6, 7, 9, 20, 30, 40, 81, 97, 98, 99, 100]
```

## variable local vs no local

```python
glb_var = 10

def product(n):
  glb_var = glb_var * n
  return glb_var

print(product(50))

---------------------------------------------------------------------------
UnboundLocalError                         Traceback (most recent call last)
<ipython-input-46-f4f209cbde33> in <module>()
      5   return glb_var
      6 
----> 7 print(product(50))
      8 

<ipython-input-46-f4f209cbde33> in product(n)
      2 
      3 def product(n):
----> 4   glb_var = glb_var * n
      5   return glb_var
      6 

UnboundLocalError: local variable 'glb_var' referenced before assignment
```

```python
# To fix the error then declare the variable as global in the function 
glb_var = 10

def product(n):
  global glb_var
  glb_var = glb_var * n
  return glb_var

print(product(50))

Output
500
```

escenario 2 variable no local. no cambia la variable global value

```python
#Don't change the value of the global variable. Try to use the nonlocal 
glb_var = 10

def outer_func(n):
  glb_var = 5
  def product(n):
    nonlocal glb_var
    glb_var = glb_var * n
  product(n)
  return glb_var

print(outer_func(50))
print(glb_var)

250
10
```

##94 extraer imagenes de un PDF USANDO PyMuPDF

```python
import fitz
doc = fitz.open("file.pdf")
for i in range(len(doc)):
    for img in doc.getPageImageList(i):
        xref = img[0]
        pix = fitz.Pixmap(doc, xref)
        if pix.n < 5:       # this is GRAY or RGB
            pix.writePNG("p%s-%s.png" % (i, xref))
        else:               # CMYK: convert to RGB first
            pix1 = fitz.Pixmap(fitz.csRGB, pix)
            pix1.writePNG("p%s-%s.png" % (i, xref))
            pix1 = None
        pix = None
```

## MERGE O FUSIONANDO PDF FILES

```python
from PyPDF2 import PdfFileMerger

merger = PdfFileMerger()

input1 = open("document1.pdf", "rb")
input2 = open("document2.pdf", "rb")
input3 = open("document3.pdf", "rb")

# add the first 3 pages of input1 document to output
merger.append(fileobj = input1, pages = (0,3))

# insert the first page of input2 into the output beginning after the second page
merger.merge(position = 2, fileobj = input2, pages = (0,1))

# append entire input3 document to the end of the output document
merger.append(input3)

# Write to an output PDF document
output = open("document-output.pdf", "wb")
merger.write(output)
```
